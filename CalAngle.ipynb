```python
def caldis(num1,num2):
    dis = distance(num1,num2)
    if dis > (6.0):
        if math.fabs(float(A[(num1)-1][6]) - float(A[(num2)-1][6])) > 20:
            if float(A[(num1)-1][6]) > float(A[(num2)-1][6]):
                dis = distanceZa(num1,num2)
                if dis > (6.0):
                    if float(A[(num1)-1][5]) > float(A[(num2)-1][5]):
                        dis = distanceZaa(num1,num2)
                    else:
                        dis = distanceZab(num1,num2)
            else:
                dis = distanceZb(num1,num2)
                if dis > (6.0):
                    if float(A[(num1)-1][5]) > float(A[(num2)-1][5]):
                        dis = distanceZba(num1,num2)
                    else:
                        dis = distanceZbb(num1,num2)            
        elif math.fabs(float(A[(num1)-1][5]) - float(A[(num2)-1][5])) > 20:
            if float(A[(num1)-1][5]) > float(A[(num2)-1][5]):
                dis = distanceYa(num1,num2)
                if dis > (6.0):
                    if float(A[(num1)-1][5]) > float(A[(num2)-1][5]):
                        dis = distanceZaa(num1,num2)
                    else:
                        dis = distanceZba(num1,num2)
            else:
                dis = distanceYb(num1,num2)
                if dis > (6.0):
                    if float(A[(num1)-1][5]) > float(A[(num2)-1][5]):
                        dis = distanceZab(num1,num2)
                    else:
                        dis = distanceZbb(num1,num2)
    return(dis);
```

```python
def angle(time):
    da = 0
    db = 0
    dc = 0
    da = caldis(int(C[time][2]),int(C[time][3]))
    db = caldis(int(C[time][2]),int(C[time][4]))
    dc = caldis(int(C[time][3]),int(C[time][4]))
    result = math.degrees(math.acos((da**2.0+dc**2.0-db**2.0)/(2.0*da*dc)))
    return(result);
```
