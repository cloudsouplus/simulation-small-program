```python
#read file
with open('data.LayerFeO','r') as f:
    data = f.read()
lines = data.split('\n')
base_atom_ids = []         #down base fixed layer
ther_atom_ids = []         #middle thermostat layer
inte_atom_ids = []         #upper interface layer
is_atoms_section = False
blank_counter = 0
for line in lines:
    
    if is_atoms_section:
        if line:
            vals = line.split()
            z = float(vals[6])
            if z < 5:
                base_atom_ids.append(int(vals[0]))
            elif z >= 5 and z < 20:
                ther_atom_ids.append(int(vals[0]))
            else:
                inte_atom_ids.append(int(vals[0]))
        else:
            blank_counter += 1
            if 1 < blank_counter:
                is_atoms_section = False
                blank_counter = 0
                
    if line == "Atoms":
        is_atoms_section = True
        
print(len(base_atom_ids))
print(len(ther_atom_ids))
print(len(inte_atom_ids))
print(vals)
```

```python
base_atom_ids = set(base_atom_ids)
ther_atom_ids = set(ther_atom_ids)
inte_atom_ids = set(inte_atom_ids)
```
