```python
def distance(num1,num2):
    a = int(num1)-1
    b = int(num2)-1
    result = math.sqrt(numpy.square(float(A[a][4]) - float(A[b][4])) + numpy.square(float(A[a][5]) - float(A[b][5])) + numpy.square(float(A[a][6]) - float(A[b][6])))
    return(result);
```

```python
def distanceZa(num1,num2):
    a = int(num1)-1
    b = int(num2)-1
    result = math.sqrt(numpy.square(float(A[a][4]) - float(A[b][4])) + numpy.square(float(A[a][5]) - float(A[b][5])) + numpy.square(float(A[a][6]) - 41.16 - float(A[b][6])))
    return(result);
def distanceZb(num1,num2):
    a = int(num1)-1
    b = int(num2)-1
    result = math.sqrt(numpy.square(float(A[a][4]) - float(A[b][4])) + numpy.square(float(A[a][5]) - float(A[b][5])) + numpy.square(float(A[a][6]) + 41.16 - float(A[b][6])))
    return(result);
```
